1. Create a New Branch It's always best practice to save new changes to a "branch" (a separate copy) instead of directly to your main branch.

```python
# Creates a new branch called "add-initial-files" and switches to it
!git checkout -b add-initial-files
```

2. Add All Your New Files This tells Git to track all the new files you just added. The . means "everything in this folder."

```python
!git add .
```

3. "Commit" Your Files (Create a Save Point) This takes all the added files and saves them as a snapshot with a message describing what you did.

```python
!git config --global user.email "jamesfarrelly@gmail.com"
!git commit -m "Add initial project files"
```

4. "Push" Your Files to GitHub (Upload) This is the final step that actually uploads your new branch and your committed files to GitHub.com.

```python
%%writefile start.sh
#!/bin/bash
# Switches to main, pulls latest code, and asks for a new branch name
git checkout main
git pull
echo "-----------------------------------"
read -p "Enter a name for your new branch (e.g., update-contact-page): " branch_name
git checkout -b $branch_name
echo "-----------------------------------"
echo "✅ Done. You are now on your new branch: $branch_name"
```

```python
%%writefile save.sh
#!/bin/bash
# Adds all files, asks for a commit message, and pushes to GitHub
git add .
echo "-----------------------------------"
read -p "Enter a short commit message (e.g., 'Fixed typo on homepage'): " commit_message
git commit -m "$commit_message"
echo "-----------------------------------"
# This next command gets your current branch name automatically
git push -u origin $(git rev-parse --abbrev-ref HEAD)
echo "-----------------------------------"
echo "✅ Done. Your changes are pushed. Now go to GitHub to open a Pull Request!"
```

```python
%%writefile cleanup.sh
#!/bin/bash
# Switches to main and pulls the latest merged code
git checkout main
git pull
echo "-----------------------------------"
echo "✅ Done. Your main branch is now up to date."
```
